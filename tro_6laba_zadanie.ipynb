```python
import matplotlib.pyplot as plt 
import numpy as np 
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = [float(i) for i in range(-50, 50)]
y = [float(i) for i in range(-33, 66)]
X, Y = np.meshgrid(x, y)
print(X)
print(Y)
Z = (np.sqrt(np.arctan((X) ** 2 + (Y) ** 2) + 3) * np.cos(np.sqrt(np.tan(((X) ** 2 + (Y) ** 2)) + 5)))
print(Z)
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')


ax.plot_surface(X, Y, Z)


plt.show()
```

```python
x = [float(i) for i in range(-50, 50)]
y = [float(i) for i in range(-33, 66)]
X, Y = np.meshgrid(x, y)
print(X)
print(Y)
Z = ((np.sqrt(Y) ** 2) + 3) * (np.cos(((X) ** 2 + (Y) ** 2)) + 5)
print(Z)
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')


ax.plot_surface(X, Y, Z, cmap='viridis')


plt.show()
```
